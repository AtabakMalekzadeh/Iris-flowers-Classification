```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC 
```

```python
models = {
    "RandomForestClassifier":RandomForestClassifier(),
    "KNN":KNeighborsClassifier(),
    "svm":SVC(),
    "tree":DecisionTreeClassifier(),
    "linear model":LogisticRegression(max_iter=3000),
    "linear svc":LinearSVC()
}
```

```python
x = df.drop("Species",axis = 1)
y = df["Species"]
```

```python
np.random.seed(23)
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
x_train.drop("Id",axis = 1,inplace=True)
x_test.drop("Id",axis = 1,inplace=True)
```

```python
results = {}
for name,model in models.items():
    model.fit(x_train,y_train)
    acc = model.score(x_test,y_test) *100
    results[name] = model.score(x_test,y_test)
    print("{} accuracy: {}%".format(name,round(acc,2)))
```

```python
bar1 = pd.DataFrame(results.values(),results.keys(),columns=["accuracy"])
```

```python
bar1.index
```

```python
labels = bar1.index
values = bar1["accuracy"]

plt.bar(labels,values)
```

```python
fig, ax = plt.subplots(figsize = (10,6))

colors = ["purple","pink","fuchsia"]

bar = ax.bar(labels,values,width=0.4,color = colors)

ax.set(xlabel = "models",
       ylabel = "accuracy")
ax.set_title("each model accuracy in bar plot", pad = 20, size = 18)
plt.xticks(rotation = 90)
plt.show()
```

```python
for r in np.linspace(1,10,10):
    model = KNeighborsClassifier(n_neighbors=int(r))
    model.fit(x_train,y_train)
    acc = model.score(x_test,y_test) * 100
    print("number of neighbors is {} and accuracy is: {}".format(r,round(acc,2)))
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
model = KNeighborsClassifier(n_neighbors=5)
model.fit(x_train,y_train)
```

```python
y_preds = model.predict(x_test)
y_preds
```

```python
scores = {
    "precision_score":precision_score(y_test,y_preds,average="micro"),
    "recall":recall_score(y_test,y_preds,average="micro"),
    "f1":f1_score(y_test,y_preds,average="micro")
}

for i,j in scores.items():
    
    print("{} : {}%".format(i,j*100))
    
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
cm = confusion_matrix(y_test,y_preds)
disp = ConfusionMatrixDisplay(confusion_matrix = cm)

disp.plot();
```
